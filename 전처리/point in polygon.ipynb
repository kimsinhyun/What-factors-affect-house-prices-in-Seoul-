# Point in Polygon
## 해당 주피터 파일은 서울의 행정동 경계구역 geojson파일과 좌표값을 통해 소속된 행정동을 찾는 파트이다
### 이렇게 하는 이유는 행정동, 법정동이 년도에 따라 이름이 조금씩 바뀌고 있고 데이터마다 행정동, 법정동의 이름이 조금씩 다르기 때문에 전처리에 어려움이 있어 본인은 Point in Polygon 방식을 택하였다

```python
gdf = gpd.GeoDataFrame.from_file('../data/seoul_geo.geojson', encoding='utf-8')

#다른 환경에서 아래 코드를 돌릴 때 pyProj오류가 발생할 수 있으므로 아래 코드를 주석해제하고 실행
import os


#-------------------------------------------------------주의!!--------------------------------------------
#C:/ .... / Anaconda3 사이의 주소는 본인 PC에 맞게 설정
os.environ['PROJ_LIB']=r"C:/Users/dblab/Anaconda3/Library/share"  
#-------------------------------------------------------주의!!--------------------------------------------

gdf.crs = {'init':'epsg:4326'}
gdf = gdf.to_crs({'init':'epsg:5179'})

#소속된 경계구역을 반환한다.
def return_office_dong(point):
    dong_df = pd.DataFrame()
    dong_df['dong'] = gdf['adm_nm'].str.split().str[2]
    dong_list = dong_df.dong.tolist()
    dong_list
    for i,dong in enumerate(dong_list):
        if(point.within(gdf.geometry[i].buffer(0)) == True):
            print(str(dong_list[i]))
            return str(dong_list[i])
    else:
        return None
# Point(126.989574,37.569006).within(gdf.geometry[0].buffer(0))
```

```python
import requests
import json

headers = {
    #아래 줄 "----" 안에 자신의 카카오맵 API key 입력
    
}
n_th_api = 0
def KakaoMap_get_lat_lon(address):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(address)
    places = requests.get(url, headers = headers).json()['documents']
    try:
        lon = places[0]['x']
        lat = places[0]['y']
    except:
        lon = 0
        lat = 0
    global n_th_api
    n_th_api += 1
    print(n_th_api, "th search lat:",lat,"lon:", lon)
    return  pd.Series([lat,lon])
```

# 20년도 주택 데이터의 소속 행정동 구하기

```python
서울주택매매20년도_geocoding = pd.read_csv('../2020년주택매매데이터(geocoding)/2020년주택매매데이터.csv').drop('Unnamed: 0',axis=1)
서울주택매매20년도_geocoding['lon'] = 서울주택매매20년도_geocoding['lon'].astype(float)
서울주택매매20년도_geocoding['lat'] = 서울주택매매20년도_geocoding['lat'].astype(float)
서울주택매매20년도_geocoding['geometry'] = 서울주택매매20년도_geocoding.apply(lambda row : Point([row['lon'], row['lat']]), axis=1)
서울주택매매20년도_geocoding = gpd.GeoDataFrame(서울주택매매20년도_geocoding, geometry='geometry')
서울주택매매20년도_geocoding.crs = {'init':'epsg:4326'}
서울주택매매20년도_geocoding = 서울주택매매20년도_geocoding.to_crs({'init':'epsg:5179'})

서울주택매매20년도_geocoding['행정동'] = 서울주택매매20년도_geocoding['geometry'].apply(return_office_dong)
```

```python
서울주택매매20년도_geocoding = 서울주택매매20년도_geocoding[~서울주택매매20년도_geocoding['행정동'].isnull()].reset_index(drop = True)
서울주택매매20년도_geocoding
```

# 지하철역

```python
with open('../data/서울시_지하철역_공간정보.geojson',encoding='utf-8') as file:
    stations = json.loads(file.read())
    file.close()
    
lineList = []
for i in range(0,len(stations['features'])):
    lineList.append(stations['features'][i]['properties']['line'])
lineList_df = pd.DataFrame(lineList)
nameList=[]
for i in range(0,len(stations['features'])):
    nameList.append(stations['features'][i]['properties']['name'])
nameList_df = pd.DataFrame(nameList)
latList = []
for i in range(0,len(stations['features'])):
    latList.append(stations['features'][i]['properties']['lat'])
latList_df = pd.DataFrame(latList)
lonList=[]
for i in range(0,len(stations['features'])):
    lonList.append(stations['features'][i]['properties']['lon'])
lonList_df = pd.DataFrame(lonList)
stations_df = pd.concat([lineList_df,nameList_df,latList_df,lonList_df],axis=1)
stations_df.columns = ['line','name','Y좌표','X좌표']
stations_df['geometry'] = stations_df.apply(lambda row : Point([row['X좌표'], row['Y좌표']]), axis=1)
stations_df = gpd.GeoDataFrame(stations_df, geometry='geometry')
stations_df.crs = {'init':'epsg:4326'}
stations_df = stations_df.to_crs({'init':'epsg:5179'})

stations_df['행정동'] = stations_df['geometry'].apply(return_office_dong)
stations_df.to_csv("../point_Polygon_전처리후_데이터/서울시지하철역.csv",encoding='utf-8-sig')
```
