```python
data=pd.read_csv('../data/15~20년도_주택매매데이터.csv')
data['행정구'] = data['시군구'].apply(lambda x: str(x).split()[1])
data['법정동'] = data['시군구'].apply(lambda x: str(x).split()[2])
data['계약년월'] = data['계약년월'].astype(int)
data['계약_년'] = data['계약년월'].apply(lambda x: int(x/100))
data['계약_월'] = data['계약년월'].apply(lambda x: int(x%100))
data['검색키워드1'] = data['시군구'] + " "+data['도로명']
data['검색키워드2'] = data['시군구'] + " "+data['단지명']
data
```

```python
temp0_5mil = data[50001:100000].reset_index(drop=True)
temp0_5mil
```

```python
import requests
import json

headers = {
    "Authorization": "_____"
}
n_th_api = 0
def KakaoMap_get_lat_lon(address):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(address)
    places = requests.get(url, headers = headers).json()['documents']
    try:
        lon = places[0]['x']
        lat = places[0]['y']
    except:
        lon = 0
        lat = 0
    global n_th_api
    n_th_api += 1
    print(n_th_api, "th search lat:",lat,"lon:", lon)
    return  pd.Series([lat,lon])
temp0_5mil[['lat','lon']]= temp0_5mil['검색키워드1'].apply(KakaoMap_get_lat_lon)
temp0_5mil
```

```python
temp0_5mil.to_csv('../주택매매데이터나누기/5~10만 데이터.csv',encoding='utf-8-sig')
```
