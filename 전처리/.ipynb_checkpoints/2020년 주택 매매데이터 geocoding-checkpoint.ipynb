```python
import pandas as pd
import numpy as np



import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
plt.rcParams['font.family'] = 'NanumGothic'




import folium
import json
from folium.plugins import MarkerCluster
import geopandas as gpd
from shapely.geometry import Point, Polygon,MultiPolygon
from pyproj import Proj




import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../주택매매_공공데이터/아파트(매매)__실거래가_20211008130144.csv',header=15)
data['행정구'] = data['시군구'].apply(lambda x: str(x).split()[1])
data['법정동'] = data['시군구'].apply(lambda x: str(x).split()[2])
data['계약년월'] = data['계약년월'].astype(int)
data['계약_년'] = data['계약년월'].apply(lambda x: int(x/100))
data['계약_월'] = data['계약년월'].apply(lambda x: int(x%100))
data['검색키워드1'] = data['시군구'] + " "+data['도로명']
data['검색키워드2'] = data['시군구'] + " "+data['단지명']
data
```

```python
import requests
import json

headers = {
    #아래 줄 "----" 안에 자신의 카카오맵 API key 입력
#     "Authorization": "--------------------"
        "Authorization": "_____"
}
n_th_api = 0
def KakaoMap_get_lat_lon(address):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(address)
    places = requests.get(url, headers = headers).json()['documents']
    try:
        lon = places[0]['x']
        lat = places[0]['y']
    except:
        lon = 0
        lat = 0
    global n_th_api
    n_th_api += 1
    print(n_th_api, "th search lat:",lat,"lon:", lon)
    return  pd.Series([lat,lon])
data[['lat','lon']]= data['검색키워드1'].apply(KakaoMap_get_lat_lon)
data
```

# 1차 geocoding(검색 키워드1 사용)

```python
#1차 geocoding 결측치 제거 후 재가공
data_completed = data[(data['lat'] != 0) | (data['lon'] != 0)].reset_index(drop=True)
data_completed
```

```python
data_missing_lon_lat = data[(data['lat'] == 0) & (data['lon'] == 0)].reset_index(drop=True)
data_missing_lon_lat
```

# 2차 geocoding(검색키워드2 사용)
- 1차 geocoding에서 좌표값을 얻어오지 못한 데이터들은 검색키워드2로 재검색

```python
#결측치 재가공
data_missing_lon_lat[['lat','lon']]= data_missing_lon_lat['검색키워드2'].apply(KakaoMap_get_lat_lon)
data_missing_lon_lat
```

```python
#2차 검색을 거쳐도 좌표값을 얻지 못하는 데이터들은 제거
data_missing_lon_lat_completed = data_missing_lon_lat[data_missing_lon_lat['lon']!=0].reset_index(drop=True)
```

```python
data_completed_final = pd.concat([data_completed,data_missing_lon_lat_completed],axis=0).reset_index(drop=True)
data_completed_final
```

```python
data_completed_final = data_completed_final[data_completed_final['lat']!=0]
data_completed_final
```

```python
data_completed_final.to_csv('../2020년주택매매데이터(geocoding)/2020년주택매매데이터.csv',encoding='utf-8-sig')
```
